```python
import numpy as np
import matplotlib.pyplot as plt

# Create a unit step signal
def unit_step_sequence(n):
    return 1 if n >= 0 else 0

# Generate samples
n_values = np.arange(-10, 11)
samples = [unit_step_sequence(n) for n in n_values]

# Create a wave
 #wave = thinkdsp.Wave(samples=samples, framerate=1)

# Plot the unit step signal
plt.step(n_values, samples, where='post')
plt.title('Unit Step Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.ylim(-0.1, 1.1)
plt.show()
```
